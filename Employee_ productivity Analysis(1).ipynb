```python
#objective 5 : Create a heatmap showing correlation among numeric variables.
numeric_cols = data.select_dtypes(include=[np.number]).columns
numeric_data = data[numeric_cols]

correlation_matrix = numeric_data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, 
            annot=True,           
            cmap='coolwarm',      
            center=0,             
            square=True,        
            fmt='.2f')           
plt.title('Correlation Heatmap of Numeric Variables')
plt.tight_layout()
plt.show()
```

```python
#Objective 6 :  Is there a significant correlation between:

#   i. Weekly_Hours and Stress_Level?

#   ii. Stress_Level and Productivity_Score?

#   iii. Remote_Work (%) and Satisfaction_Score

correlation_pairs = [
    ('Weekly_Hours', 'Stress_Level', 'Weekly Hours', 'Stress Level'),
    ('Stress_Level', 'Productivity_Score', 'Stress Level', 'Productivity Score'),
    ('Remote_Work (%)', 'Satisfaction_Score', 'Remote Work (%)', 'Satisfaction Score')
]

# Setting up the plot
plt.figure(figsize=(15, 5))
sns.set_style('whitegrid')

# Calculating and visualizing correlations
for i, (var1, var2, label1, label2) in enumerate(correlation_pairs, 1):
    corr_coef, p_value = stats.pearsonr(data[var1], data[var2])
    
    # Plotting scatter plot
    
    plt.subplot(1, 3, i)
    sns.scatterplot(x=data[var1], y=data[var2], alpha=0.6)
    plt.title(f'{label1} vs {label2}\nCorr: {corr_coef:.3f}, P-value: {p_value:.4f}', fontsize=12)
    plt.xlabel(label1, fontsize=10)
    plt.ylabel(label2, fontsize=10)
    
    # Interpreting significance
    alpha = 0.05
    print(f"\nCorrelation Analysis: {label1} vs {label2}")
    print(f"Pearson Correlation Coefficient: {corr_coef:.3f}")
    print(f"P-value: {p_value:.4f}")
    if p_value < alpha:
        print("Result: Reject the null hypothesis. There is a statistically significant correlation.")
    else:
        print("Result: Fail to reject the null hypothesis. There is no statistically significant correlation.")

plt.tight_layout()
plt.show()
```

```python
#Objective 7 : Perform an independent t-test to assess if males and females differ significantly in:

#    i. Productivity_Score
#    ii. Stress_Level

test_variables = [
    ('Productivity_Score', 'Productivity Score'),
    ('Stress_Level', 'Stress Level')
]

# Setting up the plot       
plt.figure(figsize=(12, 5))
sns.set_style('whitegrid')

# Pt-tests and visualizing
for i, (var, label) in enumerate(test_variables, 1):
    # Separating data by gender
    male_data = data[data['Gender'] == 'M'][var]
    female_data = data[data['Gender'] == 'F'][var]
    
    #  independent t-test
    t_stat, p_value = stats.ttest_ind(male_data, female_data, equal_var=False)  # Welch's t-test
    
    #  boxplot
    plt.rcParams['text.color'] = 'white' 
    plt.subplot(1, 2, i)
    sns.boxplot(x='Gender', y=var, data=data, palette='Set2')
    plt.title(f'{label} by Gender\nT-stat: {t_stat:.3f}, P-value: {p_value:.4f}', fontsize=12)
    plt.xlabel('Gender', fontsize=10)
    plt.ylabel(label, fontsize=10)
    
    #  results
    print(f"\nIndependent T-Test: {label} by Gender")
    print(f"T-statistic: {t_stat:.3f}")
    print(f"P-value: {p_value:.4f}")
    alpha = 0.05
    if p_value < alpha:
        print("Result: Reject the null hypothesis. There is a statistically significant difference between males and females.")
    else:
        print("Result: Fail to reject the null hypothesis. There is no statistically significant difference between males and females.")

plt.tight_layout()
plt.show()
```

```python
# Objective 8 : Conduct a one-way ANOVA to determine whether the mean productivity score differs across departments.

# Grouping data by department
departments = data['Department'].unique()
dept_groups = [data[data['Department'] == dept]['Productivity_Score'] for dept in departments]

f_stat, p_value = stats.f_oneway(*dept_groups)

print("One-Way ANOVA Results:")
print(f"F-statistic: {f_stat:.3f}")
print(f"P-value: {p_value:.4f}")
```

```python
# Objective 9 : Create:

#   i. A histogram for Satisfaction_Score
plt.figure(figsize=(10, 6))
plt.xlabel("Satisfaction_Score")
plt.title("A histogram for Satisfaction_Score")
sns.histplot(data["Satisfaction_Score"])

#   ii. A scatter plot showing the relationship between Stress_Level and Productivity_Score
plt.figure(figsize=(10, 6))
plt.xlabel("Productivity_Score")
plt.ylabel("Stress_Level")
plt.title("Relationship between Stress_Level and Productivity_Score")
sns.scatterplot(x = 'Productivity_Score', y='Stress_Level', data = data)

#   iii. A bar chart comparing average productivity per department
plt.figure(figsize=(10, 6))
plt.xlabel("Department")
plt.ylabel("Productivity_Score")
plt.title("Average productivity per department")
sns.barplot(x = 'Department', y='Productivity_Score', data = data)

#   iv. A violin plot of Productivity_Score by Gender
plt.figure(figsize=(10, 6))
plt.xlabel("Gender")
plt.ylabel("Productivity_Score")
plt.title("Productivity_Score by Gender")
sns.violinplot(x = 'Gender', y='Productivity_Score', data = data, hue='Gender')
```
